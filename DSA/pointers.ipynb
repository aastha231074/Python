```python
num1 = 11
num2 = num1
```

```python
print("Before num2 value is updated:")
print("num1 =", num1)
print("num2 =", num2)

print("\nnum1 points to:", id(num1))
print("num2 points to:", id(num2))
```

```python
num2 = 22
```

```python
print("After num2 value is updated:")
print("num1 =", num1)
print("num2 =", num2)

print("\nnum1 points to:", id(num1))
print("num2 points to:", id(num2))
```

# What happened ? 
Before: </br>
num1 -> 11 </br>
num2 ^ 

After: </br>
num1 -> 11 </br>
num2 -> 22 

Why? Because integers are immutable, which means they can not be changed. You can point num1 to a different integer that is stored somewhere else, but you can't chnage the number 11 once you create it.          

```python
dict1 = {'value': 11 }
dict2 = dict1

# dict1 and dict2 points to the same location in memory 
```

```python
print("Before dict2 value is updated:")
print("dict1 =", dict1)
print("dict2 =", dict2)

print("\ndict1 points to:", id(dict1))
print("dict2 points to:", id(dict2))
```

```python
dict2['value'] = 22
```

```python
print("After dict2 value is updated:")
print("dict1 =", dict1)
print("dict2 =", dict2)

print("\ndict1 points to:", id(dict1))
print("dict2 points to:", id(dict2))
```

# What happened ? 

Why? Because dictionary are mutable, which means they can be changed. Dictionary 1 and dictionary 2 continue pointing to the same address but we changed the value
